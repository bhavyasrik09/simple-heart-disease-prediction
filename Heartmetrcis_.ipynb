```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
dt = pd.read_csv('/content/heart.csv')
```

```python
dt.info()
```

```python
dt.drop_duplicates(inplace = True)
```

```python
dt
```

```python
dt.corr()
```

```python
x = dt.drop(['trestbps','chol', 'fbs', 'restecg', 'target'], axis = 1)
y = dt['target']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain, ytest = train_test_split(x,y, test_size = 0.25)
```

```python
sns.pairplot(data = dt, hue = 'fbs')
```

```python
sns.pairplot(data = dt)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 10000)
lr = lr.fit(xtrain, ytrain)
ypred = lr.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(ytest,ypred)
```

```python
a =confusion_matrix(ytest,ypred)
a
```

```python
tp = a[0][0]
tn = a[1][1]
fp = a[1][0]
fn = a[0][1]
```

```python
acc = (tp+tn)/(tp+tn+fp+fn)
```

```python
acc
```

```python
roc_auc_score(ytest, ypred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = KNeighborsClassifier(n_neighbors= 5)
```

```python
k.fit(xtrain,ytrain)
ypred = k.predict(xtest)
print(accuracy_score(ytest,ypred))
print(roc_auc_score(ytest,ypred))
print(confusion_matrix(ytest, ypred))
```
